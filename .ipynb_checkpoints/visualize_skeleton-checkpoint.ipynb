```python
def plt_base(s_np, idxs, s_pandas, fontsize, seg_synapse_df, bins, bin_offset = 0):
    radii = gaussian_filter(s_np[idxs,4] * 8, 1)
    plt.fill_between(xs * 8/1000, radii, y2 = -radii, color = config.section_colors['axon'], zorder=0, alpha = 0.5, edgecolor = 'none')
    
    mito_radius = gaussian_filter(np.sqrt( s_pandas['mito_CA'].to_numpy()[idxs] / np.pi ), 1)
    mito_labels = measure.label(mito_radius > 5)
    for mito_label in np.unique(mito_labels[mito_labels>0]):
        this_mito_idxs = np.where( mito_label == mito_labels )[0]
        this_mito_idxs = np.append( this_mito_idxs[0]-1, this_mito_idxs)
        if this_mito_idxs[-1] < len(mito_labels)-1:
            this_mito_idxs = np.append( this_mito_idxs, this_mito_idxs[-1]+1)
        color = np.ones(3) * 0/255 if mito_label == 2 else np.ones(3) * 100/255
        rad = mito_radius[this_mito_idxs] * 8
        plt.fill_between(xs[this_mito_idxs] * 8/1000, rad, y2 = -rad, color = color, zorder=1)
    seg_synapse_coords = seg_synapse_df[['x','y','z']].to_numpy()
    presynapse_xs = []
    postsynapse_xs = []
    for i_synapse in range(len(seg_synapse_df)):
        up_idx, up_dist = utils.get_up_idx_dist(s_np, seg_synapse_coords[i_synapse])
        if up_idx in idxs:
            this_x = xs[ np.where(up_idx == idxs)[0][0] ] + up_dist
            if this_x < np.inf:
                if seg_synapse_df.iloc[i_synapse]['type'] == 'pre': 
                    presynapse_xs.append( this_x )
                else: 
                    postsynapse_xs.append( this_x )
    for this_bin in bins:
        plt.plot( (8.47156 + this_bin) * np.ones(2), 750 + np.array([0,500])-bin_offset, color = 'k', linewidth=0.25)
        plt.plot( (6.52192 - this_bin) * np.ones(2), 750 + np.array([0,500])-bin_offset, color = 'k', linewidth=0.25)
    plt.plot( [8.47156,8.47156+bins[-1]], 750*np.ones(2), color = 'k', linewidth=0.25)
    plt.plot( [6.52192,6.52192-bins[-1]], 750*np.ones(2), color = 'k', linewidth=0.25)
    return presynapse_xs, postsynapse_xs

i_neuron = 848
leaf_node = 563
arbor = 'axon'

bodyId, neuron_type = neuron_quality_np[i_neuron,[0,1]]
synapse_file = home_dir + f'/saved_synapse_df/{neuron_type}_{bodyId}_synapse_df.csv'
synapse_df = pd.read_csv(synapse_file)
s_pandas = pd.read_csv(home_dir + f'/saved_clean_skeletons/s_pandas_{bodyId}_{neuron_type}_200nm.csv')
node_classes = s_pandas['node_classes'].to_numpy()
mito_radius = np.sqrt( s_pandas['mito_CA'].to_numpy() / np.pi )

in_arbor = node_classes == node_class_dict[arbor]
s_np = s_pandas.to_numpy()
leaf_nodes, branch_nodes = utils.find_leaves_and_branches(s_np)
leaf_nodes = np.array(leaf_nodes)
idxs = utils.get_down_idxs(s_np, leaf_node, ~in_arbor)[70:155]# #[45:]
xs = np.cumsum(s_np[idxs,6]) - s_np[idxs[0],6]

synapse_idxs = utils.find_closest_idxs(s_np, synapse_df)
seg_synapse_df = synapse_df.iloc[ np.where( np.isin(synapse_idxs, idxs) )[0] ]

fig_w = 3
fig_h = 0.4
for hist_type in ['presynapse', 'postsynapse', 'branch', 'mito']:
    if hist_type == 'mito':
        bins = [0, 0.5, 1.2, 2.5, 5, 10]
        bin_offset = 500
    else:
        bins = [0,0.3,0.6,1.2,2.5]
        bin_offset = 0

    fig = plt.figure(figsize=(fig_w, fig_h))
    fontsize=8
    presynapse_xs, postsynapse_xs = plt_base(s_np, idxs, s_pandas,fontsize, seg_synapse_df, bins, bin_offset = bin_offset)

    if hist_type == 'presynapse':
        this_sign = 1
        pre_y = 1000
        for i_pre in np.argsort(presynapse_xs):
            plt.scatter(presynapse_xs[i_pre]*8/1000, pre_y + this_sign*100, color = config.synapse_colors['pre'], s = 10, edgecolors = 'none', zorder=2)
            this_sign *= -1
    elif hist_type == 'postsynapse':
        this_sign = -1
        post_y = 1000
        for i_post in np.argsort(postsynapse_xs):
            plt.scatter(postsynapse_xs[i_post]*8/1000, post_y + this_sign*100, color = config.synapse_colors['post'], s = 10, edgecolors = 'none', zorder = 2)
            this_sign *= -1
    elif hist_type == 'branch':
        is_branch = np.isin(s_np[idxs,0], branch_nodes)
        is_branch[ np.where(is_branch)[0][-1] ] = False # don't count base as a branch

        branch_y = 1000
        plt.scatter(xs[is_branch] * 8/1000, branch_y * np.ones(np.sum(is_branch)),
                    s = 30, marker = "1", color = 'k', edgecolors = 'none', linewidth=0.5)
    elif hist_type == 'mito':
        plt.plot([xs[-1]*8/1000 - 0.3, xs[-1]*8/1000], [750+300,750], color = 'k', linewidth=0.25)
        plt.plot([xs[-1]*8/1000 - 0.3, xs[-1]*8/1000], [750-300,750], color = 'k', linewidth=0.25)
        plt.plot([0,0.3], [750,750+300], color = 'k', linewidth=0.25)
        plt.plot([0,0.3], [750,750-300], color = 'k', linewidth=0.25)


    
    plt.axis('off')
    plt.xlim([0, xs[-1]*8/1000])
    dx = plt.xlim()[1] - plt.xlim()[0]
    dy_mult = dx * fig_h / fig_w
    plt.ylim([ plt.ylim()[0], plt.ylim()[0] + dy_mult*1000])
    plt.savefig(home_dir + f'/saved_figures/position_GLM/{hist_type}_distribution.svg', transparent=True, bbox_inches='tight')
    plt.show()
```
